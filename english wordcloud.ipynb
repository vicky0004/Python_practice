```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text= "vicky and Shivam"
wordcloud =WordCloud(height=500,width=500,min_font_size=50).generate(text)
plt.figure(figsize = (6,6), facecolor = 'cyan')
plt.imshow(wordcloud)
plt.axis("on")
plt.tight_layout(pad=0)
plt.savefig("english_Text.png")
plt.show()
```

```python
text="Dhoni is a great cricket player"
wordcloud=WordCloud(height=800,width=800,background_color='tomato',min_font_size=8).generate(text)
#plot the WordCloud image
plt.figure(figsize=(6,6),facecolor='cyan')
plt.imshow(wordcloud)
plt.axis("on")
plt.tight_layout(pad=0)
plt.savefig("English_text.png")
plt.show()
```
